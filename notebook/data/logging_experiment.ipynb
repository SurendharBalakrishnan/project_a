```python
import logging
import os
from datetime import datetime
```

```python
logs_dir=os.path.join(os.getcwd(), 'logs')
```

```python
os.makedirs(logs_dir, exist_ok=True)
```

```python
LOG_FILE=f"{datetime.now().strftime('%Y-%m-%d_%H-%M-%S')}.log"
```

```python
## This is a log file with today's date ( name of the log file will be created in logs directory)
LOG_FILE
```

```python
## today
## with todays date log file will be created in logs directory
## tomorrow
## In tomorrow date log file will be created in logs directory

LOG_FILE_PATH=os.path.join(logs_dir, LOG_FILE)
```

![Screenshot 2025-07-26 145738.png](<attachment:Screenshot 2025-07-26 145738.png>)

```python
## check python logging documentation for more details , refer above provided screenshot

logging.basicConfig(
    filename=LOG_FILE_PATH,
    format="[ %(asctime)s ] %(levelname)s %(name)s (line:%(lineno)d) - %(message)s",
    level=logging.INFO
)
```

```python
logger=logging.getLogger("ProjectA")
```

```python
logger.info("This is an info message one")
```
